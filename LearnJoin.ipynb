```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark_Venv_Setup") \
    .master("local[*]") \
    .getOrCreate()
from pyspark.sql.functions import col, sum, count, when
print("Spark Session Created Successfully")
print(spark)
```

```python
df_one = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .option("delimiter", ",") \
    .csv("Data/table1.csv")
```

```python
df_two = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .option("delimiter", ",") \
    .csv("/home/krishna/PysparkProject/Data/table2.csv")
```

```python
result = df_one.filter(col("status")=="SUCCESS").groupBy("cust_id").agg(sum("amount").alias("total"))\
   .join(df_two, "cust_id","left") 

result.show()
```
